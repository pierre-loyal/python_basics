```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

### Distribution Plots

```python
sns.displot(tips['total_bill'], kde = True, bins=20
            )
```

### Joint Plot

```python
sns.jointplot(x='total_bill', y='tip', data= tips, kind= 'hex')
```

```python
sns.jointplot(x='total_bill', y='tip', data= tips, kind= 'kde')
```

### Pair plot

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips, hue ='smoker', palette = 'coolwarm')
```

### Categorical Data Plots

#### Barplot

```python
sns.barplot(data=tips, x='sex', y= 'total_bill', estimator= np.std)
```

```python
sns.countplot(x='sex', data =tips)

# Value count 
```

#### Box plot

```python
sns.boxplot(x='day', y='total_bill', data=tips, palette = 'rainbow', hue='sex')
```

#### Violin Plot

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
```

#### Strip Plot

```python
sns.stripplot(x='day', y='total_bill', data=tips)
```

```python
# For different styles

#sns.set_theme (style = "whitegrid")
```

#### Swarm Plot

```python
sns.swarmplot(x='day', y='total_bill', palette='rainbow', data=tips)
```

#### Factor Plot

```python
sns.catplot( x='day', y='total_bill',data=tips, kind='bar')
```

#### Regression Plot

```python
sns.lmplot(x='total_bill', y='tip', data = tips, hue='sex', scatter_kws={'s':100}, markers=['o','v'])
```

```python
sns.lmplot( x= 'total_bill', y = 'tip', data =tips, hue='sex', col='day', 
           aspect=0.8,height=22)
```

## Matrix Plots

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
fc = flights.pivot(index='month', columns='year', values='passengers')
fig, ax = plt.subplots (figsize=(12,8))
sns.heatmap(fc, annot=True)
```
