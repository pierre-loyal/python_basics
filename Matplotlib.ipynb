```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
x = np.linspace(0, 5, 11)
y = x**2
```

```python
plt.plot(x,y)
plt.title ('Here is the title')
plt.xlabel('X label')
plt.ylabel('Y label')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

### Object Oriented Method

```python
fig = plt.figure()                  # just set an empty field


axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')


axes.plot(x,y)
```

```python
fig = plt.figure()

axe1 = fig.add_axes ([0.1, 0.1, 0.8, 0.8])
axe1.plot(x,y)
axe1.set_xlabel('X label Large')
axe1.set_ylabel('Y label Large')
axe1.set_title('Large Plot')

axe2 = fig.add_axes ([0.5, 0.2, 0.35, 0.2])
axe2.plot(y,x)
axe2.set_xlabel('X label Small')
axe2.set_ylabel('Y label Small')
axe2.set_title('Small Plot')
```

```python
fig, axes = plt.subplots()

axes.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows = 1, ncols=3)

for i in axes:
    i.plot(x,y)

plt.tight_layout()
```

```python
fig, axes = plt.subplots(nrows = 1, ncols=2)

axes[0].plot(x,y, color = 'red')
axes[0].set_title('First plot')

axes[1].plot(y,x)
axes[1].set_title('2nd plot')
```

```python
fig = plt.figure(figsize=(6,2))

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows = 1, ncols=2, sharey= True, figsize = (8,2)) # sharey 

#sharey=True: This argument means that all subplots will share the same y-axis limits. This is useful for comparing different dataset
# with the same scale on the y-axis, ensuring that the y-axis is aligned and identically scaled across all subplots for easy comparison.



axes[0].plot(x,y, color = 'red')
axes[0].set_title('First plot')

axes[1].plot(y,x)
axes[1].set_title('2nd plot')
```

```python
fig = plt.figure(figsize = (8,6))

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x,y, label='1st', color ='purple', linewidth=7, alpha=0.2, ls='-.' ) # aplha is the line transparence 
axes.plot(y,x, label='2nd', color= '#FF8C00', linestyle = '--', marker = 'x', markersize = 10) # color code, hex RGB code

axes.legend(loc=2) # 1 is right, 2 is left, according to where we want it in the plot 

```
